```python
import numpy as np

# Problem 1
# Create a 1D NumPy array of shape [1x5] with random values
array_1d = np.random.uniform(size=(5,))
print("Original 1D array:", array_1d)

# Compute the mean and standard deviation
mean_value = np.mean(array_1d)
std_dev = np.std(array_1d)
print("Mean:", mean_value)
print("Standard deviation:", std_dev)

# Reshape the array into a 2D array with 5 rows and 1 column
array_2d = array_1d.reshape(5, 1)
print("Reshaped 2D array:\n", array_2d)

# Add 5 to each element and print the result
array_added_5 = array_2d + 5
print("Array after adding 5:\n", array_added_5)

# Compute the dot product of the reshaped array with itself
dot_product = np.dot(array_2d.T, array_2d) # Use transpose for dot product of column vectors
print("Dot product:\n", dot_product)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Problem 2
# Generate x values from 0 to 100 with an increment of 0.1
x = np.arange(0, 100, 0.1)

# Compute the corresponding y values using y = sin(x)
y = np.sin(x)

# Plot the sine wave
plt.figure(figsize=(10, 6))
plt.plot(x, y)

# Add labels and title
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Sine Wave Plot")

# Save the plot as a PNG file
plt.savefig("sine_wave.png")

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create two NumPy arrays: one for x values and another for y values (quadratic function)
x = np.arange(0, 101, 1)
y = 0.5 * x**2 + 2 * x + 1

# Plot the quadratic function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = 0.5x^2 + 2x + 1', linestyle='--') # Added linestyle

# Add labels and legend
plt.xlabel("x")
plt.ylabel("y")
plt.title("Quadratic Function Plot")
plt.legend()

# Add gridlines
plt.grid(True)

# Save the plot as a PDF file
plt.savefig("quadratic_function.pdf")

# Display the plot
plt.show()
```
